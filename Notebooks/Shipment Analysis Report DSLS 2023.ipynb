## Feature Engineering

```python
df['OrderDate']= pd.to_datetime(df['OrderDate'])
df['ShippedDate']= pd.to_datetime(df['ShippedDate'])
```

```python
df = df.drop(['ShipperID', 'CustomerID', 'ShipRegion','EmployeeID','Phone','SupplierID', 'ShipVia','ShipPostalCode', 'OrderID.1', 'UnitsInStock','UnitsOnOrder','ReorderLevel','Discontinued', 'ProductID.1','Picture','CategoryID.1','CategoryID','ProductID','QuantityPerUnit', 'UnitPrice.1','ShipAddress'], axis=1)
```

```python
df['Sales'] = df['UnitPrice']*df['Quantity']*(1-df['Discount'])
```

```python
df = df.dropna()
```
